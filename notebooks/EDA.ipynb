```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

train = pd.read_csv(r'/mnt/data/traffic_data/dataset/train.csv')
test = pd.read_csv(r'/mnt/data/traffic_data/dataset/test.csv')

print("Train Shape:", train.shape)
print("Test Shape:", test.shape)
train.head()
```

## Dataset Information

```
train.info()
train.describe(include='all')
```

## Missing Values

```
missing = train.isnull().sum()
missing[missing>0]
```

## Duplicate Records

```
print("Duplicates:", train.duplicated().sum())
```

## Target Variable Analysis

```
plt.figure(figsize=(8,5))
sns.histplot(train['demand'], kde=True)
plt.title('Demand Distribution')
plt.show()
```

## Feature Engineering from Timestamp

```
train['timestamp']=pd.to_datetime(train['timestamp'])
test['timestamp']=pd.to_datetime(test['timestamp'])

train['hour']=train['timestamp'].dt.hour
train['month']=train['timestamp'].dt.month
train['dayofweek']=train['timestamp'].dt.dayofweek
train['is_weekend']=train['dayofweek'].isin([5,6]).astype(int)

test['hour']=test['timestamp'].dt.hour
test['month']=test['timestamp'].dt.month
test['dayofweek']=test['timestamp'].dt.dayofweek
test['is_weekend']=test['dayofweek'].isin([5,6]).astype(int)

train[['timestamp','hour','month','dayofweek','is_weekend']].head()
```

## Numerical Feature Distributions

```
num_cols=train.select_dtypes(include=np.number).columns

for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(train[col], kde=True)
    plt.title(col)
    plt.show()
```

## Categorical Features

```
cat_cols=train.select_dtypes(include='object').columns

for col in cat_cols:
    plt.figure(figsize=(8,4))
    train[col].value_counts().head(10).plot(kind='bar')
    plt.title(col)
    plt.show()
```

## Demand vs Weather / RoadType

```
plt.figure(figsize=(8,5))
sns.boxplot(x='Weather', y='demand', data=train)
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x='RoadType', y='demand', data=train)
plt.xticks(rotation=45)
plt.show()
```

## Peak Hour Analysis

```
hourly=train.groupby('hour')['demand'].mean()

plt.figure(figsize=(10,5))
hourly.plot()
plt.title('Average Demand by Hour')
plt.xlabel('Hour')
plt.ylabel('Demand')
plt.show()
```

## Correlation Heatmap

```
plt.figure(figsize=(10,7))
corr=train.select_dtypes(include=np.number).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

## Outlier Detection

```
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=train[col])
    plt.title(col)
    plt.show()
```

## Geohash Hotspot Analysis

```
top_geo=train.groupby('geohash')['demand'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_geo.plot(kind='bar')
plt.title('Top Geohash Traffic Hotspots')
plt.show()
```
